```python
from sklearn import datasets
import pandas as pd
%pylab inline
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
```

```python
#Reading the dataset
df = pandas.read_csv("train.csv")
```

```python
#All the attributes in the dataset
df.columns
```

```python
#Categorical variables in the dataset
X = pandas.DataFrame()
X['datetime'] = df['datetime']
X['season'] = df['season']
X['holiday'] = df['holiday']
X['workingday'] = df['workingday']
X['weather'] = df['weather']
X['temp'] = df['temp']
X['atemp'] = df['atemp']
X['humidity'] = df['humidity']
X['windspeed'] = df['windspeed']
```

```python
X = X.dropna(axis=0)
```

```python
#Test data
test_data
```

```python
#Checking null values for datetime 
test_data[test_data.datetime.isnull()]
```

#### From the above table, we can see there are no null values for datetime

```python
X = X.join(pd.get_dummies(df.holiday, prefix='holiday'))
```

```python
test_data[test_data.holiday.isnull()]
```
